### **OOP**

```python
class Wheel:
  def __init__(self, material = "wood", diameter = 1, init_angle = 0):
    # protect
    self._material = material
    # private
    self.__diameter = diameter
    # public
    self.angle = init_angle

  def __str__(self):
    # result_string = "The wheel is made of " + self.material
    result_string = f"The wheel is made of {self._material} \n angle = {self.angle}"
    return result_string
  
  def rotate(self, by_angle = 0):
    self.angle += by_angle
    return f"rotated on {self.angle}"
```

```python
wheel_1 = Wheel()
print(wheel_1)
```

```python
wheel_1.diameter = 3
wheel_1.diameter 
```

```python
wheel_2 = Wheel("metal", 2, 45)
print(wheel_2)
```

```python
# inheritance
class Wheel_son(Wheel):
  def __init__(self):
    super().__init__()
    self.number = 4
  
  def rotate(self, by_angle):
    self.angle += by_angle
    return f"rotated on {self.angle}"
```

```python
wheel_3 = Wheel_son()
```
