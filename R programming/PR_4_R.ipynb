```
# library(tidyr)
library(dplyr)

# task_1
# Get a data frame from GAZ.xlsx
gas_extraction <- read.xlsx('https://github.com/qwerty29544/RpracticeBook/blob/master/2Data/01FlatTables/GAZ.xlsx?raw=true') # sheetIndex = 1, encoding = "UTF-8"

head(gas_extraction, n = 3L)
```

```
# Now we can see that there are some problems with column names, so the solution is to rename them into English
# PS: the solution had been adopted before I've written the needful encoding
# colnames(gas_extraction) <- c("Measurement date", "Pipe pressure", "Temperature", "Gas extraction",
                              # "Condensate", "Water", "ID", "Borehole shrub", "Group")
colnames(gas_extraction) <- c("Дата_замера", "Давление_в_трубе", "Температура_установки",
            "Добыча_газа", "Конденсат", "Вода", "ID_скважины", "Куст_скважины", "Группа_скважины")

head(gas_extraction, n = 3L)
```

```
# Next task is to represent data in first column as a certain date format
# Standard R-format
Sys.Date()
# is 2021-03-06
# ??? all turned into NA
gas_extraction[1] <- sapply(gas_extraction[1], function(x) { ifelse(is.na(as.numeric(x)),
      as.character(as.Date(x, format="%d/%m/%Y")), as.character(convertToDate(x, origin = "1900-01-01"))) })

head(gas_extraction, n = 10L)
```

```
# Now we need to select all None-values and turn them into NA-s
gas_extraction <- transform(gas_extraction, Давление_в_трубе = as.numeric(Давление_в_трубе))
gas_extraction <- transform(gas_extraction, Температура_установки = as.numeric(Температура_установки))
gas_extraction <- transform(gas_extraction, Добыча_газа = as.numeric(Добыча_газа))
gas_extraction <- transform(gas_extraction, Конденсат = as.numeric(Конденсат))
gas_extraction <- transform(gas_extraction, Вода = as.numeric(Вода))

head(gas_extraction, n = 4L)
```

```
# Convert the installation temperature to Kelvin units
gas_extraction <- transform(gas_extraction, Температура_установки = Температура_установки + 273.15)

head(gas_extraction, n = 4L)
```

```
# Nominal values in R - use format "factor"
gas_extraction$ID_скважины <- factor(gas_extraction$ID_скважины)
gas_extraction$Куст_скважины <- factor(gas_extraction$Куст_скважины)
gas_extraction$Группа_скважины <- factor(gas_extraction$Группа_скважины)

head(gas_extraction, n = 4L)
```

```
# Delete all NAs
gas_extraction <- gas_extraction[complete.cases(gas_extraction), ]
head(gas_extraction, n = 6L)
```

```
# Gas production to condensate ratio
gas_extraction["Отн_газа_на_конденсат"] <- gas_extraction["Добыча_газа"] / gas_extraction["Конденсат"]
# Ratio of gas production to water production
gas_extraction["Отн_газа_на_Воду"] <- gas_extraction["Добыча_газа"] / gas_extraction["Вода"]
# Ratio of water production to condensate production
gas_extraction["Отн_Воды_на_Конденсат"] <- gas_extraction["Вода"] / gas_extraction["Конденсат"]

head(gas_extraction, n = 9L)
```

```
# Filter the measurement data and get a sample of shrub production for 2018
# Convert char dates into Date-format
getYear <- function(x) {
  return (as.numeric(format(as.Date(x),'%Y'))) 
}

gas_extraction_2018 <- filter(gas_extraction, getYear(gas_extraction$Дата_замера) == 2018)

head(gas_extraction_2018, n = 4L)
```

```
# We need "Gas extraction" and "Borehole shrub"
mining <- aggregate(gas_extraction_2018$Добыча_газа, by = list(gas_extraction_2018$Куст_скважины), FUN = sum)
colnames(mining) <- c("Куст_скважины", "Добыча_газа")
mining
```

```
# Get a subsample of measurement data from the shrub by ID = 111
ID_111 <- filter(gas_extraction, ID_скважины == 111)

head(ID_111)
```

```
# Get all borehole IDs where water production has never exceeded 2 m3 / day
water <- aggregate(gas_extraction$Вода~gas_extraction$ID_скважины, gas_extraction, FUN = max)
colnames(water) <- c("ID_скважины", "Вода")
filter(water, Вода <= 2)
```

```
# Get IDs of boreholes whose total production per day did not fall below 1000 m3/day
temp <- aggregate(Добыча_газа ~ ID_скважины, gas_extraction, min)
all_ids <- unique(gas_extraction["ID_скважины"])[[1]]
length(temp[[1]]) == length(all_ids) || length(temp[[1]]) == 0
```

```
temp[order(temp$Добыча_газа, decreasing = TRUE),][3, "Добыча_газа"]
```

```
# Print the name of the group of shrubs that was the most productive in gas production in 2018
groups <- aggregate(Добыча_газа ~ Группа_скважины, gas_extraction_2018, sum)
groups
```

```
# The most productive gas extraction group
groups[order(groups$Добыча_газа, decreasing = TRUE), ][1, "Группа_скважины"]
```

```
# Print the name of the shrub that was the most productive in water production in 2018
kusts <- aggregate(Вода ~ Куст_скважины, gas_extraction_2018, sum)
kusts
```

```
# The most productive shrub for water extraction
kusts[order(kusts$Вода, decreasing = TRUE), ][1, "Куст_скважины"]
```

```
# Print the name of the shrub in which the average ratio of oil production to water production was
# the highest for all the time of observations
right_results <- filter(gas_extraction, gas_extraction$Отн_газа_на_Воду != Inf)
kusts <- aggregate(Отн_газа_на_Воду ~ Куст_скважины, right_results, mean)
kusts
```

```
# The name of the shrub in which the average ratio of gas production to water production was the highest for all the time of observations
kusts[order(kusts$Отн_газа_на_Воду, decreasing = TRUE), ][1, "Куст_скважины"]
```
