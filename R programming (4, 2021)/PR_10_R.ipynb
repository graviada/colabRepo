```r
cars <- mtcars
head(cars)
```

```r
cars$cyl <- as.factor(cars$cyl)
cars$am <- as.factor(cars$am)

# Inspect the data
head(cars[, c("hp", "wt", "cyl", "am")], 4)
```

```r
ggplot(cars, aes(x = wt, y = hp)) + 
  geom_point(aes(color = am, size = cyl), alpha = 0.5) +
  scale_color_manual(values = c("green", "red")) + labs(title = "Пузырьковая диаграмма зависимости мощности от веса", 
subtitle= "с учетом числа цилиндров автомобиля и типа коробки передач (автоматическая или нет)", x = "Мощность (hp)", y = "Вес (wt)") + theme_classic()
```

```r
cars$hp <- type.convert(cars$hp)
cars$am <- c("Mechanic", "Authomatic")

ggplot(cars, aes(x = hp, color = am)) + geom_histogram(color = "black", fill = "brown", bins = 6) + theme_bw() + labs(title = "Gross horsepower", x = "Horsepower") + facet_grid(. ~ am)
```

```r
library(openxlsx)
library(tidyr)
library(dplyr)
```

```r
covid_19 <- read.xlsx("./data_output.xlsx", detectDates = TRUE)
head(covid_19)
```

```r
covid_19["Amount"] <- rowSums(covid_19[4:length(colnames(covid_19))])
head(covid_19)
```

```r
# Number of infected people worldwide
ggplot(mapping = aes(x = covid_19[,"Date"], y = covid_19[,"Amount"])) +
geom_smooth(colour = "black", lwd = I(0.7), method = "loess", formula = y ~ x) + 
geom_point(alpha = 0.6, colour = "red", cex = I(0.7)) + 
theme_light() + 
scale_y_continuous(n.breaks = 10, labels = scales::comma) + 
scale_x_date(date_labels = "%b %Y", breaks = function(x) seq.Date(from = min(x), 
                                                 to = max(x), 
                                                 by = "2 month"),
                             minor_breaks = function(x) seq.Date(from = min(x), 
                                                                    to = max(x), 
                                                                    by = "1 month")) + 
labs(title = "Количество заражённых во всём мире", x = "Дата", y = "Количество заражённых, человек")             
```

```r
# Russia
ggplot(mapping = aes(x = covid_19[,"Date"], y = covid_19[,"Russia"])) +
geom_smooth(colour = "black", lwd = I(0.5), method = "loess", formula = y ~ x) + 
geom_line(linetype = "solid", cex = I(0.9), color = "red") +
theme(panel.background = element_rect(fill = "#FFFFC6", size = 0.5, linetype = "solid")) +
scale_y_continuous(n.breaks = 10, labels = scales::comma) + 
scale_x_date(date_labels = "%b %Y", breaks = function(x) seq.Date(from = min(x), 
                                                 to = max(x), 
                                                 by = "2 month"),
                             minor_breaks = function(x) seq.Date(from = min(x), 
                                                                    to = max(x), 
                                                                    by = "1 month")) + 
labs(title = "Количество заражённых в России", x = "Дата", y = "Количество заражённых, человек") 
```
