```r
expression_to_function <- function(expr, vars=c("x")) {
    expr <- as.expression(expr)
    
    vars_string <- ""
    if (length(vars) == 0) {
        vars_string <- ""
    }
    else if (length(vars) == 1) {
        vars_string <- as.character(vars)
    }
    else {
        for (var in vars) {
            vars_string <- paste(vars_string, as.character(var), sep=",")
        }
        vars_string <- sub(".", "", vars_string)
    }
    
    
    return (eval(parse(text=paste("(function(", vars_string, ") {", as.character(expr), "})"))))
}
```

```r
der.f_1 <- expression_to_function(f_1.der)

plot(x, der.f_1(x), type="l", main = "The derivative of the f_1(x) function")
```

```r
hyperboloid <- function(x, y, a = 1, b = 1, c = 1) sqrt(c^2 * ((x^2/a^2) + (y^2/b^2) - 1))
x <- seq(-5, 5, 0.1)
y <- seq(-5, 5, 0.1)

z_1 <- outer(x, y, hyperboloid)

persp(x, y, z_1)
persp(x, y, -z_1)
```

```r
# task 8
x_args <- seq(-5,5, 0.1)
y_args <- seq(-5,5, 0.1)

torus <- function (x, y, r=3, R=4) sqrt(-x^2-y^2-R^2-r^2+4*R^2*(x^2+y^2))

z_torus <- outer(x_args, y_args, torus)

persp(x_args, y_args ,z_torus, col = "yellow")
persp(x_args, y_args, -z_torus, col = "yellow")
```

```r
# Using plot3D - torus can be defined parametrically
par(mar = c(2, 2, 2, 2))
par(mfrow = c(1, 1))
R <- 10
r <- 4
x <- seq(0, 2*pi, length.out = 30)
y <- seq(0, 2*pi, length.out = 30)
M <- mesh(x, y)
 
alpha <- M$x
beta <- M$y

surf3D(x = (R + r*cos(alpha)) * cos(beta),
       y = (R + r*cos(alpha)) * sin(beta),
       z = r * sin(alpha),
       bty = "b2",
       main = "Torus")
```
