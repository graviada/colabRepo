```r
# 4
n <- readline(prompt = "Enter number of elements: ")
v <- vector(mode = 'integer', length = n)
for (i in 1:n) {
  if (i%%2 == 0) v[i] = 1
}
print(v)
```

```r
# 5
n <- readline(prompt = "Enter number of elements in grades10: ")
grades10 <- vector()
grades5 <- vector()

for (i in 1:n) {
  x <- as.integer(readline())
  stopifnot(x <= 10)
  grades10 <- append(grades10, x)
}

for (i in 1:n) {
  if (grades10[i] <= 1) grades5[i] = 0
  if (grades10[i] > 1 && grades10[i] <= 4) grades5[i] = 2
  if (grades10[i] >= 4 && grades10[i] <= 5) grades5[i] = 3
  if (grades10[i] >= 6 && grades10[i] <= 7) grades5[i] = 4
  if (grades10[i] >= 8 && grades10[i] <= 10) grades5[i] = 5
}

print(grades5)
```

```r
# 6
rows <- readline(prompt = "Enter a number of rows: ")
columns <- readline(prompt = "Enter a number of columns: ")
if (rows == columns) diag(rows) else print("Impossible  to create an identity matrix with such dimensions")
```
