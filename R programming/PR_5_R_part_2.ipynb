```r
# Input data
norm_vec <- function(x) sqrt(sum(x^2))

n <- as.integer(readline(prompt = "N = "))
stopifnot(n > 1)
A <- matrix(0, n, n)
f <- rep(c(1), n)
n_iter <- as.integer(readline(prompt = "Количество итераций: "))
eps <- as.numeric(readline(prompt = "Точность eps = "))
u0 <- rnorm(n)
```

```r
# Function for interation method
SIM <- function(A, f, u0, n_iter, eps) {
  stopifnot(is.matrix(A) & length(A[1,]) == length(A[,1]))
  stopifnot(is.vector(f) & length(f) == length(A[1,]))
  stopifnot(is.vector(u0) & length(u0) == length(f))
  stopifnot(is.integer(n_iter) & n_iter %% 1 == 0 & n_iter > 0)
  stopifnot(is.numeric(eps) & eps > 0)
  
  n <- length(f)
  
  divisor <- max(A, f)
  A <- A / divisor
  f <- f / divisor
  
  B <- diag(1, n, n) - A
  
  U <- matrix(0, n_iter + 1, n) # iteration matrix
  U[1,] <- u0 # vector of zero values u0
  
  for (i in 2:(n_iter + 1)) {
    U[i,] <- B %*% U[i - 1,] + f
    if (max(abs(U[i,]  - U[i - 1,])) < eps) {
      return (U[i,])
    }
  }
  
  return (U[n_iter + 1,])
}
```

```r
# Print an answer
print(SIM(A, f, u0, n_iter, eps))
```
